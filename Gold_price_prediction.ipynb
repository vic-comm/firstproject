```python
import numpy as np
import pandas as pd
from  sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
gold_data = pd.read_csv('/content/gld_price_data.xls')
gold_data.head()
```

```python

gold_data.isnull().sum()
```

```python
gold_data.describe()
```

```python
gold_data.info()
```

```python
gold_data.tail()
```

```python
x = gold_data.drop(['GLD','Date'], axis=1)
y = gold_data['GLD']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2, test_size=0.2)
```

```python
model = RandomForestRegressor() 
```

CORRELATION

```python
correlation = gold_data.corr()
```

```python
# constrcuting a heatmap to understand the corrlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

```python
# correlation values of gld
print(correlation['GLD'])
```

```python
# check the distribution of the gold price
sns.distplot(gold_data['GLD'], color='blue')
```

```python
model.fit(x_train,y_train)
```

```python
test_data_prediction = model.predict(x_test)
```

```python
# R squared
error_score = metrics.r2_score(y_test, test_data_prediction)
print(error_score)
```

## **Comparing the actual price and predicted price using scatter plots**

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color='blue', label='Actual value')
plt.plot(test_data_prediction, color='red', label='Predicted price')
plt.title("Actual price vs Predicted price")
plt.xlabel("Number of values")
plt.ylabel('GLD price')
plt.legend()
plt.show()
```
