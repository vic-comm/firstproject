## Handling missing values 

```python
# droping the cabin column
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
# replacing the missing values in age column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
titanic_data.tail()
```

```python
titanic_data.isnull().sum()
```

```python
# finding the mode of the embarked column
titanic_data['Embarked'].mode()
```

```python
titanic_data['Embarked'].mode()[0]
```

```python
# replacing the empty spaces with the mode
titanic_data["Embarked"].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
titanic_data.isnull().sum()
```

### **Data Analysis**

```python
# getting the statisticl measures about the data
titanic_data.describe()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
titanic_data['Survived'].value_counts()
```

```python
# Encoding the categorical columns
titanic_data.drop(['Name'], axis=1)
```

```python
titanic_data.replace({'Sex':{'male':0, 'female':1,}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
x = titanic_data.drop(['Ticket', 'PassengerId', 'Survived', 'Name'], axis=1)
y = titanic_data['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2, test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_score = accuracy_score(x_train_prediction, y_train)
print(x_train_prediction)
print(f"Accuracy of our model on training data is {training_data_score}")
```

```python
x_test_prediction = model.predict(x_test)
test_data_score = accuracy_score(x_test_prediction, y_test)
print(x_test_prediction)
print(f"Accuracy of our model on test data is {test_data_score}")
```

```python
x = 3,0,22,1,0,7.25,0
x_input = np.array(x)
rx_input = x_input.reshape(1, -1)
prediction = model.predict(rx_input)
print(prediction)
```
