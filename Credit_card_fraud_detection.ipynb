## **Under_sampling**

Build a sample dataset containing similar distribution of normal transactions and fradulent transactions

Number of fradulnt transactions = 211

```python
legit_sample = legit.sample(n=211)
```

Concatenating 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting the datset into features and target
x = new_dataset.drop(['Class', 'Time'], axis=1)
y = new_dataset['Class']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=3, stratify=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
training_prediction = model.predict(x_train)
```

```python
score1 = accuracy_score(y_train, training_prediction)
print(f"Training model accuracy : {score1}")
```

```python
test_prediction = model.predict(x_test)
score2 = accuracy_score(y_test, test_prediction)
print(f"Testing model accuracy : {score2}")
```
